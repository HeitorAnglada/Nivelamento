## Ordem de Preced√™ncia
A ordem de preced√™ncia em Python define como as opera√ß√µes matem√°ticas s√£o avaliadas em express√µes que cont√™m m√∫ltiplos operadores. Entender essa ordem √© crucial para escrever express√µes que se comportem como esperado. Aqui est√° a ordem de preced√™ncia, do mais alto para o mais baixo:

1. Par√™nteses (): Usados para alterar a ordem natural de preced√™ncia.

2. Exponencia√ß√£o: Calcula pot√™ncias.

3. Multiplica√ß√£o, Divis√£o, Divis√£o Inteira e M√≥dulo: Avaliados da esquerda para a direita.

4. Adi√ß√£o e Subtra√ß√£o: Tamb√©m avaliados da esquerda para a direita.

###Exemplo

````python
# Par√™nteses alteram a ordem natural
expressao = (1 + 2) * (3 - 4) / 5
print("Resultado com par√™nteses:", expressao)

# Exponencia√ß√£o tem a maior preced√™ncia
expressao = 2 ** 3 * 4
print("Resultado com exponencia√ß√£o:", expressao)  # 2**3 calculado primeiro, ent√£o multiplicado por 4

# Multiplica√ß√£o e divis√£o s√£o avaliadas antes de adi√ß√£o e subtra√ß√£o
expressao = 2 * 3 - 4 / 2
print("Resultado com multiplica√ß√£o e divis√£o:", expressao)  # 2*3 calculado, ent√£o 4/2, e finalmente 6-2

# Divis√£o inteira e m√≥dulo
expressao = 10 // 3 + 10 % 3
print("Resultado com divis√£o inteira e m√≥dulo:", expressao)  # 10//3 calculado, ent√£o 10%3, e finalmente 3+1

# A ordem da esquerda para a direita √© aplicada para operadores de mesma preced√™ncia
expressao = 5 - 2 + 3
print("Resultado com adi√ß√£o e subtra√ß√£o:", expressao)  # Calculado da esquerda para a direita, 5-2=3, ent√£o 3+3=6

# Combina√ß√£o de operadores
expressao = 5 + 2 * 3 ** 2
print("Resultado combinado:", expressao)  # 3**2 calculado primeiro, ent√£o 2*9, e finalmente 5+18
````

# Operadores L√≥gicos
Os operadores l√≥gicos em Python s√£o fundamentais para a constru√ß√£o de express√µes l√≥gicas, permitindo que voc√™ combine ou modifique condi√ß√µes booleanas. Eles s√£o essenciais para controlar o fluxo de programas atrav√©s de estruturas condicionais e de repeti√ß√£o. Existem tr√™s operadores l√≥gicos principais: and, or, e not.

| Operador | Descri√ß√£o           | Exemplo de Uso | Resultado |
|----------|---------------------|----------------|-----------|
| `or`     | 	Retorna True se pelo menos uma condi√ß√£o for verdadeira            | `True or False`       | `True`    |
| `and`     | 	Retorna True se ambas as condi√ß√µes forem verdadeiras        | `True and False`       | `False`   |

## Operador AND
O operador and retorna True se ambas as condi√ß√µes que ele une forem verdadeiras. Caso contr√°rio, retorna False.


###Exemplo

````python
idade = 25
tem_cartao = True

if idade >= 18 and tem_cartao:
    print("Acesso permitido.")
else:
    print("Acesso negado.")
````

## Operador OR
O operador or retorna True se pelo menos uma das condi√ß√µes que ele une for verdadeira. S√≥ retorna False se ambas as condi√ß√µes forem falsas.

###Exemplo

````python
chuva = False
sol = True

if chuva or sol:
    print("Pode haver um arco-√≠ris.")
else:
    print("N√£o vai ter arco-√≠ris.")
````

## Operador NOT
O operador not inverte o valor booleano de uma condi√ß√£o: transforma True em False e vice-versa.

###Exemplo

````python
fim_de_semana = False

if not fim_de_semana:
    print("Dia de trabalho.")
else:
    print("Dia de descanso.")
````

##Combinando Operadores L√≥gicos
Voc√™ pode combinar esses operadores para criar condi√ß√µes mais complexas:

###Exemplo

````python
tem_ingresso = True
fila_grande = False

if tem_ingresso and not fila_grande:
    print("Entrar no cinema.")
else:
    print("N√£o entrar no cinema.")
````

##Preced√™ncia dos Operadores L√≥gicos
A preced√™ncia define a ordem em que as opera√ß√µes s√£o avaliadas. Em Python, a preced√™ncia dos operadores l√≥gicos √©: not, and, or. Voc√™ pode usar par√™nteses para alterar a ordem de avalia√ß√£o e tornar a express√£o mais clara.

````python
a = True
b = False
c = True

resultado = a and (b or c)
print(resultado)  # True, porque b or c √© avaliado primeiro, resultando em True, e ent√£o a and True √© True.
````

#Os operadores de compara√ß√£o
Os operadores de compara√ß√£o em Python s√£o usados para comparar dois valores e retornar um valor booleano (True ou False) com base na compara√ß√£o. Eles s√£o fundamentais para controlar o fluxo de execu√ß√£o em estruturas condicionais e loops. Aqui est√° um guia detalhado sobre os operadores de compara√ß√£o dispon√≠veis em Python:



##Operadores
##Igual a (==)

##Menor que (<)

## Match case
A estrutura match case foi introduzida no Python 3.10 como uma forma mais poderosa e flex√≠vel de lidar com compara√ß√µes m√∫ltiplas, semelhante ao switch case encontrado em outras linguagens de programa√ß√£o. Ela permite que voc√™ compare uma express√£o com v√°rios poss√≠veis valores e execute um bloco de c√≥digo correspondente ao primeiro valor que combina. Isso √© particularmente √∫til para simplificar longas cadeias de if-elif-else.

### Estrutura B√°sica do match case

```python
match valor:
    case padrao1:
        # Bloco de c√≥digo para padrao1
    case padrao2:
        # Bloco de c√≥digo para padrao2
    case _:
        # Bloco de c√≥digo padr√£o (como o 'else' em if-elif-else)
```



####Exemplo

````python
comando = input("Digite um comando (start/stop/pause): ")

match comando:
    case "start":
        print("Jogo iniciado!")
    case "stop":
        print("Jogo parado.")
    case "pause":
        print("Jogo pausado...")
    case _:
        print("Comando desconhecido.")
````

### Caracter√≠sticas Importantes
**Padr√µes Literais:** Voc√™ pode combinar diretamente com strings, n√∫meros e outros valores literais.

**Vari√°veis:** Se o padr√£o for uma vari√°vel, ela combinar√° com qualquer coisa e o valor ser√° atribu√≠do a ela.

**Desestrutura√ß√£o:** Voc√™ pode desestruturar sequ√™ncias (como listas ou tuplas) diretamente no padr√£o.

**Padr√µes com Guardas:** √â poss√≠vel adicionar uma condi√ß√£o extra a um padr√£o usando a palavra-chave if.

````python
ponto = (2, 3)

match ponto:
    case (0, 0):
        print("Origem")
    case (x, 0):
        print(f"X={x}, no eixo X")
    case (0, y):
        print(f"Y={y}, no eixo Y")
    case (x, y) if x == y:
        print("Ponto na linha Y=X")
    case (x, y):
        print(f"Ponto em ({x}, {y})")
````

## Controle de Loop
Al√©m da condi√ß√£o para terminar o loop, voc√™ pode controlar a execu√ß√£o do loop while usando as instru√ß√µes break e continue.

- break: Encerra o loop imediatamente, independentemente da condi√ß√£o do while.
- continue: Pula o restante do c√≥digo dentro do loop e inicia a pr√≥xima itera√ß√£o.

## Pr√© C√°lculo

**C√°lculo de ra√≠zes reais de um polin√¥mio de segundo grau (equa√ß√£o quadr√°tica) usando a f√≥rmula de Bhaskara**

Vamos criar um c√≥digo em Python que calcula as ra√≠zes reais de um polin√¥mio de segundo grau (equa√ß√£o quadr√°tica) usando a f√≥rmula de Bhaskara. Isso pode ser √∫til para estudantes de Engenharia que est√£o praticando pr√©-c√°lculo e resolvendo equa√ß√µes quadr√°ticas.

A f√≥rmula de Bhaskara √© dada por:

$$x =  \frac{-b \pm \sqrt{(b^{2} - 4ac)} }{2a} $$

onde:

$x$ s√£o as ra√≠zes do polin√¥mio

$a$, $b$ e $c$ s√£o os coeficientes do polin√¥mio de segundo grau $(ax^2 + bx + c)$

Vamos ao c√≥digo:

````python
import math

# Entrada dos coeficientes
a = float(input("Digite o coeficiente 'a' do polin√¥mio de segundo grau: "))
b = float(input("Digite o coeficiente 'b' do polin√¥mio de segundo grau: "))
c = float(input("Digite o coeficiente 'c' do polin√¥mio de segundo grau: "))

# C√°lculo do delta
delta = b**2 - 4*a*c

# Verifica√ß√£o das ra√≠zes
if delta < 0:
    print("O polin√¥mio n√£o possui ra√≠zes reais.")
elif delta == 0:
    x = -b / (2*a)
    print("A raiz do polin√¥mio √©:", x)
else:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print("As ra√≠zes do polin√¥mio s√£o:", x1, "e", x2)
````

## F√≠sica

**C√°lculo do lan√ßamento de um proj√©til, considerando a velocidade inicial, √¢ngulo de lan√ßamento e a acelera√ß√£o da gravidade**

Para o assunto de F√≠sica, podemos criar um exemplo de c√≥digo em Python que calcula o lan√ßamento de um proj√©til, considerando a velocidade inicial, √¢ngulo de lan√ßamento e a acelera√ß√£o da gravidade. Esse tipo de c√°lculo √© comum em problemas de mec√¢nica, e a seguir, vamos desenvolver um programa simples para realizar esse c√°lculo.

Vamos utilizar as equa√ß√µes do movimento de um proj√©til em duas dimens√µes:

Posi√ß√£o no eixo x: $$x = v_{0} * cos(Œ∏) * t$$

Posi√ß√£o no eixo y: $$y = v_{0} * sen(Œ∏) * t - (1/2) * g * t^2$$

Onde:

$v_{0}$ √© a velocidade inicial do proj√©til em m/s,

$Œ∏$ √© o √¢ngulo de lan√ßamento em graus,

$t$ √© o tempo de voo em segundos,

$g$ √© a acelera√ß√£o da gravidade (aproximadamente 9.81 $m/s^2$).

Vamos ao c√≥digo:


````python
import math

# Entrada dos dados
v0 = float(input("Digite a velocidade inicial do proj√©til (m/s): "))
theta = float(input("Digite o √¢ngulo de lan√ßamento do proj√©til (graus): "))

# Convers√£o de graus para radianos
theta_rad = math.radians(theta)

# C√°lculos
g = 9.81
alcance_maximo = (v0**2 * math.sin(2*theta_rad)) / g
altura_maxima = (v0**2 * math.sin(theta_rad)**2) / (2*g)

# Exibi√ß√£o dos resultados
print("Alcance m√°ximo:", alcance_maximo, "metros")
print("Altura m√°xima:", altura_maxima, "metros")
````

Seja $y(t)$ a fun√ß√£o que descreve o sinal ECG:

$y(t) = 0.5 \cdot \sin(2 \pi \cdot 1.5 \cdot t) + 0.3 \cdot \sin(2 \pi \cdot 2.5 \cdot t)$

- $t$ √© o tempo.

- $sin$ representa a fun√ß√£o seno.

- $2\pi$ √© uma constante relacionada ao per√≠odo da fun√ß√£o seno.

**Benchmark de diferentes algoritmos de ordena√ß√£o**

Para os cursos de Engenharia e Ci√™ncia da computa√ß√£o usaremos um exemplo de medi√ß√£o de desempenho (benchmark) de diferentes algoritmos de ordena√ß√£o. No campo da Engenharia da Computa√ß√£o e da Ci√™ncia da Computa√ß√£o, a ordena√ß√£o de dados √© uma opera√ß√£o fundamental para organizar informa√ß√µes em uma sequ√™ncia espec√≠fica. Veremos como implementar os algoritmos de ordena√ß√£o Bubble Sort e Selection Sort, e como medir o tempo de execu√ß√£o de cada um deles. Isso √© relacionado √† arquitetura e organiza√ß√£o de computadores, j√° que diferentes algoritmos de ordena√ß√£o podem ter impactos significativos no desempenho, dependendo da arquitetura do computador.

Antes de irmos ao c√≥digo, √© necess√°rio fazermos uma breve explica√ß√£o sobre algoritmos de ordena√ß√£o. OS Algoritmos de Ordena√ß√£o s√£o procedimentos ou conjuntos de regras que permitem organizar um conjunto de dados em uma determinada ordem, como ordem crescente ou decrescente. Eles s√£o utilizados para simplificar a busca e a an√°lise de informa√ß√µes dentro desses dados. Existem diferentes tipos de algoritmos de ordena√ß√£o, desde os simples, como Bubble Sort e Selection Sort, at√© os mais complexos e eficientes, como Merge Sort e Quick Sort. A escolha do algoritmo de ordena√ß√£o adequado depende do tamanho dos dados, da efici√™ncia desejada e do contexto de uso.

Neste exemplo, usaremos dois algoritmos de ordena√ß√£o:
1.   **Bubble Sort**: O Bubble Sort √© um algoritmo simples de ordena√ß√£o. Ele funciona comparando elementos adjacentes na lista e trocando-os se estiverem na ordem errada. Ele percorre repetidamente a lista, comparando pares de elementos adjacentes e trocando-os at√© que a lista esteja completamente ordenada. Esse processo de "borbulhar" os elementos maiores ou menores para suas posi√ß√µes corretas √© repetido at√© que nenhum swap seja necess√°rio.
2.   **Selection Sort**: O Selection Sort √© outro algoritmo de ordena√ß√£o simples. Ele divide a lista em duas partes: a parte ordenada e a parte n√£o ordenada. O algoritmo encontra o menor (ou maior, dependendo se voc√™ est√° ordenando em ordem crescente ou decrescente) elemento na parte n√£o ordenada e o coloca no final da parte ordenada. Esse processo √© repetido at√© que todos os elementos estejam na parte ordenada.

![alt text](https://lh5.googleusercontent.com/rKq4GXLXEFhpjH-bVgDtovKUKsLR2ClizRqFiGz4979nY1X3q-bqA3L8o_Ah9nbYq--ZEwKe9Pn4XVPk74Ec-ybieiC2k6CfqVZ8_u5T6bw4T86JgvhY5i8pzGfDp-gbEdWM1tfjRbf4QuiKq4rkixY)

Bubble Sort Simplificado


````python
arr = [64, 34, 25, 12, 22, 11, 90]  # Lista de exemplo

# Algoritmo Bubble Sort
n = len(arr)
for i in range(n):
    for j in range(0, n-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print("Array ordenado pelo Bubble Sort:", arr)
````

Selection Sort Simplificado


````python
arr = [64, 25, 12, 22, 11]  # Lista de exemplo

# Algoritmo Selection Sort
n = len(arr)
for i in range(n):
    min_idx = i
    for j in range(i+1, n):
        if arr[j] < arr[min_idx]:
            min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]

print("Array ordenado pelo Selection Sort:", arr)
````

## Engenharias El√©trica e de Energia

**C√°lculo do Fator de Pot√™ncia**

Para falar sobre fator de pot√™ncia, primeiramente vamos entender sobre as diferentes pot√™ncias em Correntes Alternadas(CA), que s√£o 3, pot√™ncia Ativa, Reativa e Aparente.

*Pot√™ncia Ativa (P):*
√â a pot√™ncia real que realiza trabalho ou, em outras palavras, a pot√™ncia que √© convertida em energia √∫til no circuito. Esta √© a pot√™ncia que √© efetivamente usada em dispositivos el√©tricos para produzir um efeito √∫til (como luz, calor, movimento, etc.). Ela √© medida em watts ($W$).

*Pot√™ncia Reativa (Q):*
√â a pot√™ncia que est√° continuamente sendo armazenada e liberada pelo campo magn√©tico de indutores e pelo campo el√©trico de capacitores presentes no circuito. A pot√™ncia reativa n√£o realiza um trabalho √∫til, mas √© necess√°ria para manter a tens√£o e a corrente em fase em circuitos de corrente alternada (CA) que possuem elementos reativos. Ela √© medida em volt-amperes reativos ($VAr$).


*Pot√™ncia Aparente (S):*
Representa a magnitude total da pot√™ncia em um circuito de CA e √© a combina√ß√£o vetorial da pot√™ncia ativa e reativa. √â a pot√™ncia "bruta" que circula no circuito, parte dela realiza trabalho √∫til (pot√™ncia ativa) e outra parte √© armazenada e liberada (pot√™ncia reativa). Ela √© medida em volt-amperes (VA).

Abaixo temos o tri√¢ngulo das pot√™ncias, que √© uma representa√ß√£o gr√°fica da rela√ß√£o entre essas pot√™ncias, o √¢ngulo Fi($œÜ$) √© o √¢ngulo de fase, que representa a diferen√ßa entre a tens√£o e a corrente no circuito, quanto menor esse √¢ngulo menor ser√° o fator de pot√™ncia.

![alt_text](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSzyAWIAIa551YwxLcI3flXDyS9rB6TrE-fMxR9h_0JM-5rkhSPJYkBG283NQd11KICebI&usqp=CAU)


O fator de pot√™ncia √© a raz√£o entre a pot√™ncia ativa e a potencia aparente, serve para medir a propor√ß√£o de energia que est√° sendo convertida em trabalho √∫til em um circuito, seu valor varia de 0 a 1. Sua form√∫la √© a seguinte:

$$FP = \frac{P}{S}$$
ou tamb√©m:
$$FP = cos(œÜ)$$

Onde:
1.  $FP$ √© o fator de pot√™ncia.
2.  $P$ √© a pot√™ncia ativa.
3.  $S$ √© a pot√™ncia aparente.
4.  $œÜ$ √© o √¢ngulo de fase.


Agora vamos ao c√≥digo:

````python
import math

# Entrada de dados
tensao = float(input("Digite a tens√£o (V): "))
corrente = float(input("Digite a corrente (A): "))
angulo_fase = float(input("Digite o √¢ngulo de fase entre tens√£o e corrente (graus): "))

# C√°lculos das pot√™ncias
potencia_aparente = tensao * corrente
potencia_ativa = potencia_aparente * math.cos(math.radians(angulo_fase))
potencia_reativa = potencia_aparente * math.sin(math.radians(angulo_fase))

# C√°lculo do fator de pot√™ncia
fator_potencia = potencia_ativa / potencia_aparente

# Exibi√ß√£o dos resultados
print("Pot√™ncia Ativa: ", potencia_ativa, "W")
print("Pot√™ncia Reativa: ", potencia_reativa, "VAr")
print("Pot√™ncia Aparente: ", potencia_aparente, "VA")
print("Fator de Pot√™ncia: ", fator_potencia)
````

### B√¥nus



**Encontrando um caminho de menor dist√¢ncia**

Um exemplo √∫til na Engenharia Ferrovi√°ria e Log√≠stica √© utilizar um c√≥digo para calcular a menor dist√¢ncia entre dois lugares. Para isso podemos utilizar o algoritmo de Dijkstra num grafo que represente os pontos e as suas conex√µes. A seguir um grafo de exemplo, os v√©rtices de A a E s√£o os objetivos e as arestas s√£o as liga√ß√µes entre eles com suas respectivas dist√¢ncias.

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/a/a8/Prim_Algorithm_0.svg/250px-Prim_Algorithm_0.svg.png)

Com o algoritmo de Djikstra √© possivel calcular a menor dist√¢ncia entre o n√≥ de origem(o ponto inicial) e todos os outros n√≥s do grafo. Ele funciona da seguinte forma:


1.   Escolha um n√≥ de origem;
2.   Defina as dist√¢ncias como 0 para o n√≥ de origem e infinito para os outros n√≥s;
3.  Visita os n√≥s adjacentes(conectados) ao n√≥ de origem e salva a dist√¢ncia entre eles.
4.  Salva o n√≥ de origme como visitado;
5.  Escolha um n√≥ n√£o visitado com a menor dist√¢ncia da origem;
6.  Visite os n√≥s adjacentes a partir dele;
7.  Salva a dist√¢ncia min√≠ma entre calculada a partir do n√≥ atual e a que j√° estava salva;
8.  Marca o n√≥ como visitado;
9.  Repita os passos de 5 a 8 at√© ter visitado todos os n√≥s; *texto em it√°lico*

Para explica√ß√£o mais detalhada sobre o algoritmo visite: <https://www.freecodecamp.org/portuguese/news/algoritmo-de-caminho-de-custo-minimo-de-dijkstra-uma-introducao-detalhada-e-visual/>

Vamos ao c√≥digo, para esse exemplo usaremos o grafo apresentado anteriormente.

````python
# Exemplo de grafo representado como um dicion√°rio de dicion√°rios
grafo = {
    'A': {'B': 7, 'D': 5},
    'B': {'A': 7, 'C': 8, 'D': 9, 'E': 7},
    'C': {'B': 8, 'E': 5},
    'D': {'A': 5, 'B': 9, 'E': 15, 'F': 6},
    'E': {'B': 7, 'C': 5, 'D': 15, 'F': 8, 'G': 9},
    'F': {'D': 6, 'E': 8, 'G': 11},
    'G': {'E': 9, 'F': 11}
}

# Inicializa√ß√£o das estruturas de dados
distancias = {no: float('inf') for no in grafo}  # inicializa a dist√¢ncia de todos os n√≥s como infinito
visitados = set()

# Solicita√ß√£o do n√≥ de in√≠cio ao usu√°rio
no_inicio = input("Digite o n√≥ de inicio: ")
if no_inicio not in grafo:
    print("O n√≥ informado n√£o existe.")
else:
    distancias[no_inicio] = 0

    while len(visitados) < len(grafo):
        # Encontrar o n√≥ n√£o visitado mais pr√≥ximo do in√≠cio
        menor_distancia = float('inf')
        no_menor_distancia = None
        for no, distancia in distancias.items():
            if no not in visitados and distancia < menor_distancia:
                menor_distancia = distancia
                no_menor_distancia = no

        if no_menor_distancia is None:
            break

        visitados.add(no_menor_distancia)

        # Calcula a dist√¢ncia para os n√≥s vizinhos
        for vizinho, peso in grafo[no_menor_distancia].items():
            distancia = distancias[no_menor_distancia] + peso
            if distancia < distancias[vizinho]:
                distancias[vizinho] = distancia

    # Exibi√ß√£o dos resultados
    print("Caminhos mais curtos a partir de", no_inicio)
    for no, distancia in distancias.items():
        print(f"Para o n√≥ {no}, dist√¢ncia: {distancia}")
````

## Engenharia Mec√¢nica

**C√°lculo da for√ßa de arrasto de um corpo em movimento**

Um exemplo de √∫til na √°rea da Engenharia Mec√¢nica √© calcular a for√ßa de arrasto de um corpo em movimento, como um carro por exemplo.
A for√ßa de arrasto √© a for√ßa de resist√™ncia ao movimento de um objeto em um fluido, por exemplo um carro ao se mover recebe uma for√ßa de arrasto do ar ao se mover, como ilustrado na imagem abaixo.

![alt text](https://s4.static.brasilescola.uol.com.br/be/2020/02/arraste-do-ar.jpg)

A f√≥rmula para calcular a for√ßa de arrasto √© a seguinte:

$$F_D = \frac{1}{2} * \rho * v^2 * C_D * A$$

Onde:
1.  $\rho$ √© a densidade do fluido;
2.  $v$ √© a velocidade do objeto;
3.  $C_D$ √© o coeficiente de arrasto, que depende do formato do objeto;
4.  $A$ √© a √°rea transversal

Para esse exemplo vamos considerar que o objeto √© um cilindro se movendo no ar. Faremos um gr√°fico que mostre a for√ßa de arrasto conforme a velocidade do objeto aumenta, para isso utilizaremos a biblioteca `matplotlib` para fazer o gr√°fico.

Vamos ao c√≥digo:

````python
import matplotlib.pyplot as plt

# Valores para o c√°lculo da for√ßa de arrasto
coeficiente_arrasto = 0.82  # coeficiente de arrasto de um cilindro
densidade_ar = 1.225  # kg/m¬≥ (densidade do ar ao n√≠vel do mar)
area_objeto = float(input("Digite a √°rea frontal(em m¬≤) do objeto: "))  # m¬≤ (√°rea frontal do objeto)

# Lista de velocidades
velocidades = list(range(0, 100, 5))

# Lista para armazenar as for√ßas de arrasto
forcas_arrasto = []

# C√°lculo das for√ßas de arrasto para cada velocidade
for v in velocidades:
    forca_arrasto = 0.5 * coeficiente_arrasto * densidade_ar * area_objeto * v**2
    forcas_arrasto.append(forca_arrasto)

# Plotagem do gr√°fico
plt.plot(velocidades, forcas_arrasto)
plt.xlabel('Velocidade (m/s)')
plt.ylabel('For√ßa de Arrasto (N)')
plt.title('For√ßa de Arrasto em fun√ß√£o da Velocidade')
plt.grid(True)
plt.show()
````

## Engenharia de Telecomunica√ß√µes

**C√°lculo da perda de pot√™ncia no espa√ßo livre (Free Space Path Loss - FSPL)**

Um exemplo pr√°tico na √°rea de Engenharia de Telecomunica√ß√µes pode ser um programa em Python que calcula a perda de pot√™ncia no espa√ßo livre (Free Space Path Loss - FSPL) entre um transmissor e um receptor.

![alt text](https://analog.intgckts.com/wp-content/uploads/sites/2/2015/03/free-space-path-loss-wavefronts.png)

A perda de pot√™ncia no espa√ßo livre √© um conceito importante em telecomunica√ß√µes que descreve a atenua√ß√£o do sinal de radiofrequ√™ncia √† medida que se propaga pelo espa√ßo livre, sem obstru√ß√µes significativas. √â crucial para o dimensionamento e planejamento de sistemas de comunica√ß√£o sem fio.

A f√≥rmula para calcular a perda de pot√™ncia no espa√ßo livre √© dada por:

$$FSPL (dB) = 20 * log10(d) + 20 * log10(f) + 20 * log10(4 * \frac{\pi}{c})$$

Onde:

$d$ √© a dist√¢ncia entre o transmissor e o receptor em metros (m).

$f$ √© a frequ√™ncia da transmiss√£o em Hertz (Hz).

$c$ √© a velocidade da luz no v√°cuo, aproximadamente $3.0 x 10^8$ metros por segundo (m/s).

$log10$ √© a fun√ß√£o logar√≠tmica na base 10.

Vamos ao c√≥digo:

````python
import math

# Solicita√ß√£o de entrada dos dados diretamente
distancia = float(input("Digite a dist√¢ncia entre o transmissor e o receptor (em metros): "))
frequencia = float(input("Digite a frequ√™ncia da transmiss√£o (em Hertz): "))

# Verifica√ß√£o de valores positivos para dist√¢ncia e frequ√™ncia
if distancia > 0 and frequencia > 0:
    c = 3.0 * 10**8  # Velocidade da luz no v√°cuo em m/s

    # C√°lculo da perda de pot√™ncia no espa√ßo livre em dB
    fspl_db = 20 * math.log10(distancia) + 20 * math.log10(frequencia) + 20 * math.log10(4 * math.pi / c)

    # Exibi√ß√£o do resultado
    print(f"A perda de pot√™ncia no espa√ßo livre √© de {fspl_db:.2f} decib√©is (dB).")
else:
    print("Por favor, digite valores positivos para a dist√¢ncia e a frequ√™ncia.")
````
