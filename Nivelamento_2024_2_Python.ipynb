## Sintaxe Básica

Python é conhecido por sua sintaxe limpa e legível, que promove uma programação mais intuitiva. Aqui estão alguns pontos-chave:

### Indentação

Imagine que você está organizando uma pasta de arquivos importantes. Dentro dessa pasta, você tem várias subpastas, cada uma dedicada a um tipo específico de documento. A maneira como você organiza estas subpastas dentro da pasta principal é um pouco semelhante à indentação em Python.

A indentação em Python ajuda a organizar o código de maneira que possamos ver claramente qual código pertence a qual comando ou ação. Assim como as subpastas ajudam a organizar os documentos relacionados, a indentação ajuda a organizar blocos de código relacionados sob comandos como if, for, while etc.

Em Python, "indentar" significa adicionar um espaço no início de uma linha de código para mostrar que essa linha pertence a um comando anterior. É como colocar um documento na subpasta correta. Se o código não for corretamente indentado, o Python não entenderá como você quer que ele organize suas ações, assim como você teria dificuldade em encontrar documentos em uma pilha desorganizada.

### Comentários
Use o símbolo # para adicionar comentários de uma linha ao seu código e '''comentário''' para adicionar comentários sem se preocupar com o número de linhas.

- Inteiro (int): Usado para representar números inteiros.

- Booleano (bool): Usado para representar True ou False.

# Operações Matemáticas

Python suporta as operações matemáticas básicas como adição, subtração, multiplicação, e divisão, além de operações mais complexas como módulo (resto da divisão), exponenciação, e divisão inteira.

| Operador | Descrição                         | Exemplo   |
|----------|-----------------------------------|-----------|
| +        | Adição                            | 5 + 3 = 8 |
| -        | Subtração                         | 5 - 3 = 2 |
| *        | Multiplicação                     | 5 * 3 = 15|
| /        | Divisão                           | 5 / 3 ≈ 1.6667|

## Ordem de Precedência
A ordem de precedência em Python define como as operações matemáticas são avaliadas em expressões que contêm múltiplos operadores. Entender essa ordem é crucial para escrever expressões que se comportem como esperado. Aqui está a ordem de precedência, do mais alto para o mais baixo:

1. Parênteses (): Usados para alterar a ordem natural de precedência.

2. Exponenciação: Calcula potências.

3. Multiplicação, Divisão, Divisão Inteira e Módulo: Avaliados da esquerda para a direita.

4. Adição e Subtração: Também avaliados da esquerda para a direita.

###Exemplo

````python
# Parênteses alteram a ordem natural
expressao = (1 + 2) * (3 - 4) / 5
print("Resultado com parênteses:", expressao)

# Exponenciação tem a maior precedência
expressao = 2 ** 3 * 4
print("Resultado com exponenciação:", expressao)  # 2**3 calculado primeiro, então multiplicado por 4

# Multiplicação e divisão são avaliadas antes de adição e subtração
expressao = 2 * 3 - 4 / 2
print("Resultado com multiplicação e divisão:", expressao)  # 2*3 calculado, então 4/2, e finalmente 6-2

# Divisão inteira e módulo
expressao = 10 // 3 + 10 % 3
print("Resultado com divisão inteira e módulo:", expressao)  # 10//3 calculado, então 10%3, e finalmente 3+1

# A ordem da esquerda para a direita é aplicada para operadores de mesma precedência
expressao = 5 - 2 + 3
print("Resultado com adição e subtração:", expressao)  # Calculado da esquerda para a direita, 5-2=3, então 3+3=6

# Combinação de operadores
expressao = 5 + 2 * 3 ** 2
print("Resultado combinado:", expressao)  # 3**2 calculado primeiro, então 2*9, e finalmente 5+18
````

# Operadores Lógicos
Os operadores lógicos em Python são fundamentais para a construção de expressões lógicas, permitindo que você combine ou modifique condições booleanas. Eles são essenciais para controlar o fluxo de programas através de estruturas condicionais e de repetição. Existem três operadores lógicos principais: and, or, e not.

| Operador | Descrição           | Exemplo de Uso | Resultado |
|----------|---------------------|----------------|-----------|
| `or`     | 	Retorna True se pelo menos uma condição for verdadeira            | `True or False`       | `True`    |
| `and`     | 	Retorna True se ambas as condições forem verdadeiras        | `True and False`       | `False`   |
| `not`      | Inverte o valor booleano da condição	| `not False`        | `True`    |

## Operador AND
O operador and retorna True se ambas as condições que ele une forem verdadeiras. Caso contrário, retorna False.


###Exemplo

````python
idade = 25
tem_cartao = True

if idade >= 18 and tem_cartao:
    print("Acesso permitido.")
else:
    print("Acesso negado.")
````

## Operador OR
O operador or retorna True se pelo menos uma das condições que ele une for verdadeira. Só retorna False se ambas as condições forem falsas.

###Exemplo

````python
chuva = False
sol = True

if chuva or sol:
    print("Pode haver um arco-íris.")
else:
    print("Não vai ter arco-íris.")
````

## Operador NOT
O operador not inverte o valor booleano de uma condição: transforma True em False e vice-versa.

###Exemplo

````python
fim_de_semana = False

if not fim_de_semana:
    print("Dia de trabalho.")
else:
    print("Dia de descanso.")
````

##Combinando Operadores Lógicos
Você pode combinar esses operadores para criar condições mais complexas:

###Exemplo

````python
tem_ingresso = True
fila_grande = False

if tem_ingresso and not fila_grande:
    print("Entrar no cinema.")
else:
    print("Não entrar no cinema.")
````

##Precedência dos Operadores Lógicos
A precedência define a ordem em que as operações são avaliadas. Em Python, a precedência dos operadores lógicos é: not, and, or. Você pode usar parênteses para alterar a ordem de avaliação e tornar a expressão mais clara.

###Exemplo

````python
a = True
b = False
c = True

resultado = a and (b or c)
print(resultado)  # True, porque b or c é avaliado primeiro, resultando em True, e então a and True é True.
````

#Os operadores de comparação

Eles são fundamentais para controlar o fluxo de execução em estruturas condicionais e loops. Aqui está um guia detalhado sobre os operadores de comparação disponíveis em Python:

`x == y`

`x != y`

`x > y`

`x < y`

`x >= y`

`x <= y`

# Estruturas Condicionais
As estruturas condicionais em Python permitem que você execute diferentes ações com base em determinadas condições. São fundamentais para a tomada de decisão dentro do seu código.


## If

A instrução if é a mais simples das estruturas condicionais. Ela avalia uma condição e, se a condição for verdadeira (True), executa um bloco de código.

###Exemplo

````python
idade = 20
if idade >= 18:
    print("Você é maior de idade.")
````

## Elif

elif, uma abreviação de "else if", permite verificar múltiplas condições, uma após a outra, até que uma seja verdadeira.

###Exemplo

````python
idade = 16
if idade >= 18:
    print("Você é maior de idade.")
elif idade < 18:
    print("Você é menor de idade.")
````

## Else

else captura qualquer condição que não seja capturada pelos if ou elif anteriores.

###Exemplo

````python
idade = 12
if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
````

###Exemplo

````python
# Calculando a média
media = 5

# Verificando a condição da média
if media >= 10:
    print("A média dos números é alta.")
elif media >= 5:
    print("A média dos números é moderada.")
else:
    print("A média dos números é baixa.")
````

## Match case
A estrutura match case foi introduzida no Python 3.10 como uma forma mais poderosa e flexível de lidar com comparações múltiplas, semelhante ao switch case encontrado em outras linguagens de programação. Ela permite que você compare uma expressão com vários possíveis valores e execute um bloco de código correspondente ao primeiro valor que combina. Isso é particularmente útil para simplificar longas cadeias de if-elif-else.

### Estrutura Básica do match case

```python
match valor:
    case padrao1:
        # Bloco de código para padrao1
    case padrao2:
        # Bloco de código para padrao2
    case _:
        # Bloco de código padrão (como o 'else' em if-elif-else)
```



####Exemplo

````python
comando = input("Digite um comando (start/stop/pause): ")

match comando:
    case "start":
        print("Jogo iniciado!")
    case "stop":
        print("Jogo parado.")
    case "pause":
        print("Jogo pausado...")
    case _:
        print("Comando desconhecido.")
````

## Controle de Loop
Além da condição para terminar o loop, você pode controlar a execução do loop while usando as instruções break e continue.

- break: Encerra o loop imediatamente, independentemente da condição do while.
- continue: Pula o restante do código dentro do loop e inicia a próxima iteração.

###Exemplo

````python
contador = 5
while contador > 0:
    if contador == 3:
        contador -= 1
        continue  # Pula para a próxima iteração sem imprimir 3
    print(contador)
    if contador == 2:
        break  # Encerra o loop quando contador é 2
    contador -= 1
````

## Pré Cálculo

**Cálculo de raízes reais de um polinômio de segundo grau (equação quadrática) usando a fórmula de Bhaskara**

Vamos criar um código em Python que calcula as raízes reais de um polinômio de segundo grau (equação quadrática) usando a fórmula de Bhaskara. Isso pode ser útil para estudantes de Engenharia que estão praticando pré-cálculo e resolvendo equações quadráticas.

A fórmula de Bhaskara é dada por:

$$x =  \frac{-b \pm \sqrt{(b^{2} - 4ac)} }{2a} $$

onde:

$x$ são as raízes do polinômio

$a$, $b$ e $c$ são os coeficientes do polinômio de segundo grau $(ax^2 + bx + c)$

Vamos ao código:

````python
import math

# Entrada dos coeficientes
a = float(input("Digite o coeficiente 'a' do polinômio de segundo grau: "))
b = float(input("Digite o coeficiente 'b' do polinômio de segundo grau: "))
c = float(input("Digite o coeficiente 'c' do polinômio de segundo grau: "))

# Cálculo do delta
delta = b**2 - 4*a*c

# Verificação das raízes
if delta < 0:
    print("O polinômio não possui raízes reais.")
elif delta == 0:
    x = -b / (2*a)
    print("A raiz do polinômio é:", x)
else:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print("As raízes do polinômio são:", x1, "e", x2)
````

## Física

**Cálculo do lançamento de um projétil, considerando a velocidade inicial, ângulo de lançamento e a aceleração da gravidade**

Para o assunto de Física, podemos criar um exemplo de código em Python que calcula o lançamento de um projétil, considerando a velocidade inicial, ângulo de lançamento e a aceleração da gravidade. Esse tipo de cálculo é comum em problemas de mecânica, e a seguir, vamos desenvolver um programa simples para realizar esse cálculo.

Vamos utilizar as equações do movimento de um projétil em duas dimensões:

Posição no eixo x: $$x = v_{0} * cos(θ) * t$$

Posição no eixo y: $$y = v_{0} * sen(θ) * t - (1/2) * g * t^2$$

Onde:

$v_{0}$ é a velocidade inicial do projétil em m/s,

$θ$ é o ângulo de lançamento em graus,

$t$ é o tempo de voo em segundos,

$g$ é a aceleração da gravidade (aproximadamente 9.81 $m/s^2$).

Vamos ao código:


````python
import math

# Entrada dos dados
v0 = float(input("Digite a velocidade inicial do projétil (m/s): "))
theta = float(input("Digite o ângulo de lançamento do projétil (graus): "))

# Conversão de graus para radianos
theta_rad = math.radians(theta)

# Cálculos
g = 9.81
alcance_maximo = (v0**2 * math.sin(2*theta_rad)) / g
altura_maxima = (v0**2 * math.sin(theta_rad)**2) / (2*g)

# Exibição dos resultados
print("Alcance máximo:", alcance_maximo, "metros")
print("Altura máxima:", altura_maxima, "metros")
````

**Benchmark de diferentes algoritmos de ordenação**

Para os cursos de Engenharia e Ciência da computação usaremos um exemplo de medição de desempenho (benchmark) de diferentes algoritmos de ordenação. No campo da Engenharia da Computação e da Ciência da Computação, a ordenação de dados é uma operação fundamental para organizar informações em uma sequência específica. Veremos como implementar os algoritmos de ordenação Bubble Sort e Selection Sort, e como medir o tempo de execução de cada um deles. Isso é relacionado à arquitetura e organização de computadores, já que diferentes algoritmos de ordenação podem ter impactos significativos no desempenho, dependendo da arquitetura do computador.

Antes de irmos ao código, é necessário fazermos uma breve explicação sobre algoritmos de ordenação. OS Algoritmos de Ordenação são procedimentos ou conjuntos de regras que permitem organizar um conjunto de dados em uma determinada ordem, como ordem crescente ou decrescente. Eles são utilizados para simplificar a busca e a análise de informações dentro desses dados. Existem diferentes tipos de algoritmos de ordenação, desde os simples, como Bubble Sort e Selection Sort, até os mais complexos e eficientes, como Merge Sort e Quick Sort. A escolha do algoritmo de ordenação adequado depende do tamanho dos dados, da eficiência desejada e do contexto de uso.

Neste exemplo, usaremos dois algoritmos de ordenação:
1.   **Bubble Sort**: O Bubble Sort é um algoritmo simples de ordenação. Ele funciona comparando elementos adjacentes na lista e trocando-os se estiverem na ordem errada. Ele percorre repetidamente a lista, comparando pares de elementos adjacentes e trocando-os até que a lista esteja completamente ordenada. Esse processo de "borbulhar" os elementos maiores ou menores para suas posições corretas é repetido até que nenhum swap seja necessário.
2.   **Selection Sort**: O Selection Sort é outro algoritmo de ordenação simples. Ele divide a lista em duas partes: a parte ordenada e a parte não ordenada. O algoritmo encontra o menor (ou maior, dependendo se você está ordenando em ordem crescente ou decrescente) elemento na parte não ordenada e o coloca no final da parte ordenada. Esse processo é repetido até que todos os elementos estejam na parte ordenada.

![alt text](https://lh5.googleusercontent.com/rKq4GXLXEFhpjH-bVgDtovKUKsLR2ClizRqFiGz4979nY1X3q-bqA3L8o_Ah9nbYq--ZEwKe9Pn4XVPk74Ec-ybieiC2k6CfqVZ8_u5T6bw4T86JgvhY5i8pzGfDp-gbEdWM1tfjRbf4QuiKq4rkixY)

Bubble Sort Simplificado


````python
arr = [64, 34, 25, 12, 22, 11, 90]  # Lista de exemplo

# Algoritmo Bubble Sort
n = len(arr)
for i in range(n):
    for j in range(0, n-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print("Array ordenado pelo Bubble Sort:", arr)
````

Selection Sort Simplificado


````python
arr = [64, 25, 12, 22, 11]  # Lista de exemplo

# Algoritmo Selection Sort
n = len(arr)
for i in range(n):
    min_idx = i
    for j in range(i+1, n):
        if arr[j] < arr[min_idx]:
            min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]

print("Array ordenado pelo Selection Sort:", arr)
````

## Engenharias Elétrica e de Energia

**Cálculo do Fator de Potência**

Para falar sobre fator de potência, primeiramente vamos entender sobre as diferentes potências em Correntes Alternadas(CA), que são 3, potência Ativa, Reativa e Aparente.

*Potência Ativa (P):*
É a potência real que realiza trabalho ou, em outras palavras, a potência que é convertida em energia útil no circuito. Esta é a potência que é efetivamente usada em dispositivos elétricos para produzir um efeito útil (como luz, calor, movimento, etc.). Ela é medida em watts ($W$).

*Potência Reativa (Q):*
É a potência que está continuamente sendo armazenada e liberada pelo campo magnético de indutores e pelo campo elétrico de capacitores presentes no circuito. A potência reativa não realiza um trabalho útil, mas é necessária para manter a tensão e a corrente em fase em circuitos de corrente alternada (CA) que possuem elementos reativos. Ela é medida em volt-amperes reativos ($VAr$).


*Potência Aparente (S):*
Representa a magnitude total da potência em um circuito de CA e é a combinação vetorial da potência ativa e reativa. É a potência "bruta" que circula no circuito, parte dela realiza trabalho útil (potência ativa) e outra parte é armazenada e liberada (potência reativa). Ela é medida em volt-amperes (VA).

Abaixo temos o triângulo das potências, que é uma representação gráfica da relação entre essas potências, o ângulo Fi($φ$) é o ângulo de fase, que representa a diferença entre a tensão e a corrente no circuito, quanto menor esse ângulo menor será o fator de potência.

![alt_text](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSzyAWIAIa551YwxLcI3flXDyS9rB6TrE-fMxR9h_0JM-5rkhSPJYkBG283NQd11KICebI&usqp=CAU)


O fator de potência é a razão entre a potência ativa e a potencia aparente, serve para medir a proporção de energia que está sendo convertida em trabalho útil em um circuito, seu valor varia de 0 a 1. Sua formúla é a seguinte:

$$FP = \frac{P}{S}$$
ou também:
$$FP = cos(φ)$$

Onde:
1.  $FP$ é o fator de potência.
2.  $P$ é a potência ativa.
3.  $S$ é a potência aparente.
4.  $φ$ é o ângulo de fase.


Agora vamos ao código:

````python
import math

# Entrada de dados
tensao = float(input("Digite a tensão (V): "))
corrente = float(input("Digite a corrente (A): "))
angulo_fase = float(input("Digite o ângulo de fase entre tensão e corrente (graus): "))

# Cálculos das potências
potencia_aparente = tensao * corrente
potencia_ativa = potencia_aparente * math.cos(math.radians(angulo_fase))
potencia_reativa = potencia_aparente * math.sin(math.radians(angulo_fase))

# Cálculo do fator de potência
fator_potencia = potencia_ativa / potencia_aparente

# Exibição dos resultados
print("Potência Ativa: ", potencia_ativa, "W")
print("Potência Reativa: ", potencia_reativa, "VAr")
print("Potência Aparente: ", potencia_aparente, "VA")
print("Fator de Potência: ", fator_potencia)
````

### Bônus



**Encontrando um caminho de menor distância**

Um exemplo útil na Engenharia Ferroviária e Logística é utilizar um código para calcular a menor distância entre dois lugares. Para isso podemos utilizar o algoritmo de Dijkstra num grafo que represente os pontos e as suas conexões. A seguir um grafo de exemplo, os vértices de A a E são os objetivos e as arestas são as ligações entre eles com suas respectivas distâncias.

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/a/a8/Prim_Algorithm_0.svg/250px-Prim_Algorithm_0.svg.png)

Com o algoritmo de Djikstra é possivel calcular a menor distância entre o nó de origem(o ponto inicial) e todos os outros nós do grafo. Ele funciona da seguinte forma:


1.   Escolha um nó de origem;
2.   Defina as distâncias como 0 para o nó de origem e infinito para os outros nós;
3.  Visita os nós adjacentes(conectados) ao nó de origem e salva a distância entre eles.
4.  Salva o nó de origme como visitado;
5.  Escolha um nó não visitado com a menor distância da origem;
6.  Visite os nós adjacentes a partir dele;
7.  Salva a distância miníma entre calculada a partir do nó atual e a que já estava salva;
8.  Marca o nó como visitado;
9.  Repita os passos de 5 a 8 até ter visitado todos os nós; *texto em itálico*

Para explicação mais detalhada sobre o algoritmo visite: <https://www.freecodecamp.org/portuguese/news/algoritmo-de-caminho-de-custo-minimo-de-dijkstra-uma-introducao-detalhada-e-visual/>

Vamos ao código, para esse exemplo usaremos o grafo apresentado anteriormente.

````python
# Exemplo de grafo representado como um dicionário de dicionários
grafo = {
    'A': {'B': 7, 'D': 5},
    'B': {'A': 7, 'C': 8, 'D': 9, 'E': 7},
    'C': {'B': 8, 'E': 5},
    'D': {'A': 5, 'B': 9, 'E': 15, 'F': 6},
    'E': {'B': 7, 'C': 5, 'D': 15, 'F': 8, 'G': 9},
    'F': {'D': 6, 'E': 8, 'G': 11},
    'G': {'E': 9, 'F': 11}
}

# Inicialização das estruturas de dados
distancias = {no: float('inf') for no in grafo}  # inicializa a distância de todos os nós como infinito
visitados = set()

# Solicitação do nó de início ao usuário
no_inicio = input("Digite o nó de inicio: ")
if no_inicio not in grafo:
    print("O nó informado não existe.")
else:
    distancias[no_inicio] = 0

    while len(visitados) < len(grafo):
        # Encontrar o nó não visitado mais próximo do início
        menor_distancia = float('inf')
        no_menor_distancia = None
        for no, distancia in distancias.items():
            if no not in visitados and distancia < menor_distancia:
                menor_distancia = distancia
                no_menor_distancia = no

        if no_menor_distancia is None:
            break

        visitados.add(no_menor_distancia)

        # Calcula a distância para os nós vizinhos
        for vizinho, peso in grafo[no_menor_distancia].items():
            distancia = distancias[no_menor_distancia] + peso
            if distancia < distancias[vizinho]:
                distancias[vizinho] = distancia

    # Exibição dos resultados
    print("Caminhos mais curtos a partir de", no_inicio)
    for no, distancia in distancias.items():
        print(f"Para o nó {no}, distância: {distancia}")
````

## Engenharia Mecânica

**Cálculo da força de arrasto de um corpo em movimento**

Um exemplo de útil na área da Engenharia Mecânica é calcular a força de arrasto de um corpo em movimento, como um carro por exemplo.
A força de arrasto é a força de resistência ao movimento de um objeto em um fluido, por exemplo um carro ao se mover recebe uma força de arrasto do ar ao se mover, como ilustrado na imagem abaixo.

![alt text](https://s4.static.brasilescola.uol.com.br/be/2020/02/arraste-do-ar.jpg)

A fórmula para calcular a força de arrasto é a seguinte:

$$F_D = \frac{1}{2} * \rho * v^2 * C_D * A$$

Onde:
1.  $\rho$ é a densidade do fluido;
2.  $v$ é a velocidade do objeto;
3.  $C_D$ é o coeficiente de arrasto, que depende do formato do objeto;
4.  $A$ é a área transversal

Para esse exemplo vamos considerar que o objeto é um cilindro se movendo no ar. Faremos um gráfico que mostre a força de arrasto conforme a velocidade do objeto aumenta, para isso utilizaremos a biblioteca `matplotlib` para fazer o gráfico.

Vamos ao código:

````python
import matplotlib.pyplot as plt

# Valores para o cálculo da força de arrasto
coeficiente_arrasto = 0.82  # coeficiente de arrasto de um cilindro
densidade_ar = 1.225  # kg/m³ (densidade do ar ao nível do mar)
area_objeto = float(input("Digite a área frontal(em m²) do objeto: "))  # m² (área frontal do objeto)

# Lista de velocidades
velocidades = list(range(0, 100, 5))

# Lista para armazenar as forças de arrasto
forcas_arrasto = []

# Cálculo das forças de arrasto para cada velocidade
for v in velocidades:
    forca_arrasto = 0.5 * coeficiente_arrasto * densidade_ar * area_objeto * v**2
    forcas_arrasto.append(forca_arrasto)

# Plotagem do gráfico
plt.plot(velocidades, forcas_arrasto)
plt.xlabel('Velocidade (m/s)')
plt.ylabel('Força de Arrasto (N)')
plt.title('Força de Arrasto em função da Velocidade')
plt.grid(True)
plt.show()
````

## Engenharia Naval

**Cálculo da área molhada de um navio**

Uma forma de utilizar a programação na Engenharia Naval é criar um programa para calcular a área molhada de um navio, ou seja, a área da superfície do casco que está em contato com a água.

![alt text](https://pub.mdpi-res.com/jmse/jmse-08-00898/article_deploy/html/images/jmse-08-00898-g001.png?1605079377)

Para simplificação, consideraremos o casco como um cilindro truncado. Com isso, teríamos a seguinte equação para o cálculo:

$$ A = 2 \pi * (R+r) * H$$

O usuário irá inserir o raio da base maior (na linha d'água), o raio da base menor (na quilha) e a altura do casco. O programa então calculará a área molhada total. Vamos lá:


````python
import math

# Solicitação de entrada dos dados diretamente
raio_externo = float(input("Digite o raio da base maior (em metros) do cilindro truncado: "))
raio_interno = float(input("Digite o raio da base menor (em metros) do cilindro truncado: "))
altura_barco = float(input("Digite a altura (em metros) do cilindro truncado (barco): "))

# Cálculo da área de superfície molhada
A_cilindro_total = 2 * math.pi * (raio_externo + raio_interno) * altura_barco

# Exibição do resultado
print("Área de superfície molhada total do barco (em metros quadrados):", A_cilindro_total)
````

## Engenharia Sanitária e Ambiental

**Cálculo da Índice de Qualidade da Água (IQA)**

Vamos utilizar o código que calcula a concentração de uma solução para simplificar o cálculo do Índice de Qualidade da Água (IQA) para o curso de Engenharia Sanitária e Ambiental. O IQA é uma medida da qualidade da água com base em diversos parâmetros físicos, químicos e biológicos.

![alt text](https://portalinfohidro.igam.mg.gov.br/images/Figura_3.png)

Vamos considerar um exemplo simplificado em que o IQA é calculado a partir de dois parâmetros: concentração de oxigênio dissolvido (OD) e pH. Neste exemplo, não estaremos levando em conta outros parâmetros importantes para um cálculo mais preciso do IQA, mas é apenas um exemplo para fins didáticos.

Vamos ao código:

````python
# Solicitação de entrada dos dados diretamente
massa_od = float(input("Digite a massa de oxigênio dissolvido (em gramas): "))
volume_agua = float(input("Digite o volume de água (em litros): "))
ph = float(input("Digite o valor do pH: "))

# Cálculo da concentração de oxigênio dissolvido
concentracao_od = massa_od / volume_agua

# Verificação do valor de pH e cálculo do IQA
if 0 <= ph <= 14:
    # Cálculo do IQA supondo que seja a soma da concentração de OD e do pH
    iqa = concentracao_od + ph
    print(f"O Índice de Qualidade da Água (IQA) é: {iqa:.2f}")
else:
    print("Valor de pH inválido. O pH deve estar no intervalo de 0 a 14.")
````

## Engenharia de Telecomunicações

**Cálculo da perda de potência no espaço livre (Free Space Path Loss - FSPL)**

Um exemplo prático na área de Engenharia de Telecomunicações pode ser um programa em Python que calcula a perda de potência no espaço livre (Free Space Path Loss - FSPL) entre um transmissor e um receptor.

![alt text](https://analog.intgckts.com/wp-content/uploads/sites/2/2015/03/free-space-path-loss-wavefronts.png)

A perda de potência no espaço livre é um conceito importante em telecomunicações que descreve a atenuação do sinal de radiofrequência à medida que se propaga pelo espaço livre, sem obstruções significativas. É crucial para o dimensionamento e planejamento de sistemas de comunicação sem fio.

A fórmula para calcular a perda de potência no espaço livre é dada por:

$$FSPL (dB) = 20 * log10(d) + 20 * log10(f) + 20 * log10(4 * \frac{\pi}{c})$$

Onde:

$d$ é a distância entre o transmissor e o receptor em metros (m).

$f$ é a frequência da transmissão em Hertz (Hz).

$c$ é a velocidade da luz no vácuo, aproximadamente $3.0 x 10^8$ metros por segundo (m/s).

$log10$ é a função logarítmica na base 10.

Vamos ao código:

````python
import math

# Solicitação de entrada dos dados diretamente
distancia = float(input("Digite a distância entre o transmissor e o receptor (em metros): "))
frequencia = float(input("Digite a frequência da transmissão (em Hertz): "))

# Verificação de valores positivos para distância e frequência
if distancia > 0 and frequencia > 0:
    c = 3.0 * 10**8  # Velocidade da luz no vácuo em m/s

    # Cálculo da perda de potência no espaço livre em dB
    fspl_db = 20 * math.log10(distancia) + 20 * math.log10(frequencia) + 20 * math.log10(4 * math.pi / c)

    # Exibição do resultado
    print(f"A perda de potência no espaço livre é de {fspl_db:.2f} decibéis (dB).")
else:
    print("Por favor, digite valores positivos para a distância e a frequência.")
````
